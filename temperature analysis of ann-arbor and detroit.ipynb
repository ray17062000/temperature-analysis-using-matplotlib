```python
det = pd.read_csv("detroit", delimiter='\t')
det
```

```python
jan = det.Mean_Max_Temp.iloc[0]
feb = det.Mean_Max_Temp.iloc[1]
mar = det.Mean_Max_Temp.iloc[2]
apr = det.Mean_Max_Temp.iloc[3]
may = det.Mean_Max_Temp.iloc[4]
jun = det.Mean_Max_Temp.iloc[5]
jul = det.Mean_Max_Temp.iloc[6]
aug = det.Mean_Max_Temp.iloc[7]
sep = det.Mean_Max_Temp.iloc[8]
octo = det.Mean_Max_Temp.iloc[9]
nov = det.Mean_Max_Temp.iloc[10]
dec = det.Mean_Max_Temp.iloc[11]
```

```python
weather['Det_Max_Temp'] = [jan, feb, mar, apr, may, jun, jul, aug, sep, octo, nov, dec]
weather
```

```python
jan = det.Mean_Min_Temp.iloc[0]
feb = det.Mean_Min_Temp.iloc[1]
mar = det.Mean_Min_Temp.iloc[2]
apr = det.Mean_Min_Temp.iloc[3]
may = det.Mean_Min_Temp.iloc[4]
jun = det.Mean_Min_Temp.iloc[5]
jul = det.Mean_Min_Temp.iloc[6]
aug = det.Mean_Min_Temp.iloc[7]
sep = det.Mean_Min_Temp.iloc[8]
octo = det.Mean_Min_Temp.iloc[9]
nov = det.Mean_Min_Temp.iloc[10]
dec = det.Mean_Min_Temp.iloc[11]
```

```python
weather['Det_Min_Temp'] = [jan, feb, mar, apr, may, jun, jul, aug, sep, octo, nov, dec]
weather
```

```python
weather.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
weather
```

```python
x = weather.index.tolist()
y1 = weather['Ann_Max_Temp'].tolist()
y2 = weather['Ann_Min_Temp'].tolist()
y3 = weather['Det_Max_Temp'].tolist()
y4 = weather['Det_Min_Temp'].tolist()

plt.plot(x, y1, '-s', label = "Ann-Arbor Max Temp", color = 'red')
plt.plot(x, y2, '-*', label = "Ann-Arbor Min Temp", color = 'red')
plt.plot(x, y3, '-o', label = "Detroit Max Temp", color = 'blue')
plt.plot(x, y4, '-x', label = "Detroil Min Temp", color = 'blue')

plt.xlabel("MONTH")
plt.ylabel("TEMPERATURE in °F ")
plt.title('Minimum and Maximun Temp of Ann-arbor vs Detroit (1981-2010)')
 
plt.legend(frameon=False)
[plt.gca().spines[loc].set_visible(False) for loc in ['top', 'right']]

plt.show()
```
